```
import pandas as pd
pf = pd.read_csv('https://raw.githubusercontent.com/PacktWorkshops/The-Data-Science-Workshop/master/Chapter01/Dataset/dataset_44_spambase.csv')
print (pf)
target = pf.pop('class')
print (target)
from sklearn.ensemble import RandomForestClassifier
seed = 168
rf_model = RandomForestClassifier (random_state = seed)
rf_model.fit (pf, target)
preds = rf_model.predict(pf)
print(preds)
from sklearn.metrics import accuracy_score
accuracy_score (target, preds)
```
